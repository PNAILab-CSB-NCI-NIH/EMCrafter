# Full Noise Generator
This tutorial covers a wrapper class that is used to generate noise more automated based
on the NoiseShaper and NoiseModulator classes

## Load Libraries

```python
import sys
import numpy as np
from pathlib import Path
sys.path.append(str(Path.cwd()) + "/../src")

import EMCrafter
from EMCrafter.noise import NoiseGenerator
EMCrafter.base._set_logging(minimal=True)
```

## Initializing the NoiseGenerator

```python
# Parameters
apix = 0.732                       # Resoluiton in A/pixel
box_size = 480                     # Box size
image_shape = (box_size, box_size) # Image shape
```

```python
# By default, NoiseGenerator starts with a Gaussian shape
# and and a flat frequency modulation -> so *NO* MODULATION
ngen = NoiseGenerator(verbose=2)
noise_image = ngen.simulate(shape=image_shape, n=1)
_ = ngen.plot()
```

## Setting customized generation using previous tutorials

```python
# We can also simply load pickle files for each of classes: Shaper and Modulator
ngen = NoiseGenerator(verbose=2)
ngen.load_shaper("data/pickle/noise_shape.pkl")
ngen.load_modulator("data/pickle/noise_freq.pkl")
noise_image = ngen.simulate(shape=image_shape, n=1)
_ = ngen.plot()
```
