# Orientation Sampling
This tutorial covers how one can can create:
- Random orientation samples
- Random clusters of orientation

## Load Libraries

```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd()) + "/../src")

import EMCrafter
from EMCrafter.sampler import OrientationSampler
EMCrafter.base._set_logging(minimal=True)
```

## Generating random Euler angles

```python
# Verbosity level
# v = 0: Silent
# v = 1: Minimal (default)
# v = 2: Setters included
# v = 3+: More detailed
verbose = 2
```

```python
# We initialize the sampler as
osampler = OrientationSampler(verbose=verbose)
```

```python
# From where we can generate, for example, 100 random orientations:
angles = osampler.sample(n=100)

# And visualize them
fig = osampler.plot()
```

```python
# We can also generate, for example, 10 Gaussian orientations centered in their
# cluster angle positions and with a defined sigma, for each of the 50 clusters
angles = osampler.sample_clusters(  
    n = 10,             # Generate 10 random orientations per cluster
    n_clusters = 50,    # Generate 20 random clusters of orientations 
    sigma = 10)         # Sigma as 10 degrees
fig = osampler.plot()   # Visualize
```

### Storing/Loading class

```python
# We can save this sample for multiple use afterwards. For instance, this saves not only
# the sampler but also the generated angles, which can be used for multiple simulations
# in case one must compare different simulations using same sampled orientation

# Save
osampler.save("data/pickle/orientation_sampler.pkl")

# Load
osampler = OrientationSampler().load("data/pickle/orientation_sampler.pkl")
```
